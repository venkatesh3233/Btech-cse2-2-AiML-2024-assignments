```python
import numpy as np
v=np.array([1,2,3,4,5])
a=np.array([6,7,8,9,3])
m=np.multiply(a,v)
print('multiplication:',m)
s=np.divide(a,v)
print('division:',s)
print('powerof:',np.power(v,a))
text=np.array2string(v+a)
print("adddition of two=",text)
print("np.sin(v):", np.sin(v))
print("np.log(v):", np.log(v))
print("np.log2(v):", np.log2(v))
print("np.exp(v):", np.exp(v))
```

```python
import matplotlib.pyplot as plt


people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]


plt.figure(figsize=(8, 6))
plt.scatter(age, height, color='yellow', label='Height')
plt.scatter(age, weight, color='green', label='Weight')
plt.title('Scatter Plot of Age vs Health Metrics')
plt.xlabel('Age')
plt.ylabel('Health Metrics')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, width=bar_width, label='Height', color='yellow', align='center')
plt.bar(index, weight, width=bar_width, label='Weight', color='green', align='edge')
plt.title('Bar Chart of Health Metrics by Person')
plt.xlabel('People')
plt.ylabel('Health Metrics')
plt.xticks(index, people)
plt.legend()
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(height, bins=20, color='yellow', alpha=0.7, label='Height')
plt.hist(weight, bins=20, color='green', alpha=0.7, label='Weight')
plt.title('Histogram of Health Metrics')
plt.xlabel('Health Metrics')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y')
plt.show()

```
